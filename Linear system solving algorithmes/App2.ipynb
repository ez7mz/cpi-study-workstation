## Main- function to calculate the QR decomposition:

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
    *** Function to Add Rows and Columns of Matrix Id to Obtained House-Holder Matrix:
</p>

```python
def Add_Id(Matrix):
    """
    Matrix: the given matrix
    n: the matrix size
    B and D: two row vectors of Zeros
    we do the concatenation of given matrix with a row and a column of Zeros
    then we replace the diagonal element by 1
    """
    n = len(Matrix)
    B = array([zeros(n)])
    D = array([zeros(n + 1)])
    c = concatenate((B, Matrix), axis=0)
    c = concatenate((D.T, c), axis=1)
    c[0, 0] = 1
    return c
```

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
    *** Function to calculate the Householder matrix:
</p>

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
    we have :<br>
    $\left.\begin{array}{rl}v_{k}= & \frac{a_{k}}{\left\|a_{k}\right\|}-e_{1} \\ u_{k}= & \frac{v_{k}}{\left\|v_{k}\right\|}\end{array}\right\} \Rightarrow u_{k}=\frac{a_{k}-\left\|a_{k}\right\| e_{1}}{\left\|a_{k}-\right\| a_{k}\left\|e_{1}\right\|}$
</p>

```python
def HS(mat):
    """
    mat: this is the given matrix
    n: the matrix size
    Id: the identity matrix of size n
    e1: first canonical basis vector
    u: a vector obtained by the above method
    H: this is the Hous-Holder matrix
    HA: the H*A product
    """
    n = len(mat)
    Id = eye(n)

    e1 = array([Id[:, 0]])
    a = array([mat[:, 0]])
    v = a - linalg.norm(a, 2) * e1
    u = v / linalg.norm(v, 2)

    H = eye(n) - 2 * transpose(u).dot(u)
    HA = H.dot(mat)

    return H, HA
```

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
    *** Function to calculate the QR Decomposition:
</p>

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
   With : <br>
        The QR factorization of A is a factorization of the form $A=Q R$ with
$Q \in M_{n \times n}(R)$ orthogonal and $R \in M_{n \times n}(R)$ upper triangular. <br>
    such as : $R=H_{n-1} H_{n-2} \cdots H_{1} A$&nbsp;&nbsp;&nbsp;and&nbsp;&nbsp;&nbsp;$Q=\left(H_{n-1} H_{n-2} \cdots H_{1}\right)^{-1}$
</p>

```python
def QR_decomp(A):
    """
    A: the given matrix
    Hmax: list to store Hous-Holder matrices
    HAmax: list to store the H*A matrices
    M: is a temporary Matrix
    @: operator for the matrix product
    R: the QR decomposition matrix R (return value)
    Q: the QR decomposition matrix Q (rerun value)
    """
    n = len(A)
    Hmax = []
    HAmax = []

    H1, HA1 = HS(A)
    M = HA1[1:, 1:]
    Hmax.append(H1)
    HAmax.append(HA1)
    
    for i in range(1, n - 1):
        H, HA = HS(M)
        HAmax.append(HA)
        M = HA[1:, 1:]
        k = i
        while k > 0:
            H = Add_Id(H)
            k -= 1
        Hmax.append(H)

    R = eye(n)
    for i in range(n - 1):
        R = Hmax[i] @ R

    Q = transpose(R)
    R = R.dot(A)

    return Q, R
```

```python
from numpy import *
```

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
   Given $A$, we form the sequences of matrices $A_{i}, Q_{i}, R_{i}$ as follows:
    <br>
</p>

<p style="font-family: consolas, serif; font-size:11pt; font-style:italic">
   $\left\{\begin{array}{l}A_{0}=A \\ \operatorname{for} i=0,1, \ldots \\ A_{i}=Q_{i} R_{i} \\ A_{i+1}=R_{i} Q_{i}\end{array}\right.$
    <br>
</p>

```python
def valeurP_avec_QR(A, Maxiter=500):
    """
    A: the given matrix
    Maxiter: maximum number of iterations
    val_Propre: list to store the eigenvalues
    next_A: A at the next iteration
    the last loop is to extract the diagonal elements of the matrix A
    at the nth iteration
    """
    val_Propre = []
    next_A = A
    for i in range(Maxiter):
        Q, R = QR_decomp(next_A)
        next_A = R @ Q

    print("The matrix A at iteration K:\n")
    print(next_A)
    print()

    for i in range(len(A)):
        val_Propre.append(next_A[i, i])

    return val_Propre
```

```python
A = array([[4, 4, 2, 0], [4, 5, 0, 0], [2, 0, 6, 1], [0, 0, 1, 2]])

valeur = valeurP_avec_QR(A)

print("The eigenvalues of the matrix A (initial) are:\n")
print(valeur)
```

    Solving the system A*X = B using the QR method
    <br>
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    Principle : <br>
        $\left\{\begin{array}{l}A x=B \\ A=QR\end{array} \Rightarrow Q \cdot R*x = B \Rightarrow\left\{\begin{array}{l}Q y=B \\ R x=y\end{array}\right.\right.$
    <br>
</p>

```python
from numpy import *
```

```python
A = array([[4, 4, 2, 0], [4, 5, 0, 0], [2, 0, 6, 1], [0, 0, 1, 2]])
B = [1, -1, 0, 1]
```

```python
Q, R = QR_decomp(A)
```

```python
Y = linalg.solve(Q, B)
X = linalg.solve(R, Y)
```

```python
print("\nThe solution of the system is :\n")
print("X = ", X)
print()
```

## EXERCICE - Q3:

<p style="font-family: consolas, serif; font-size:14pt; font-style:italic">
    Solving the system A*X = B using the gradient method
    <br>
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    Principle : <br>
        We consider the following problem: minimize on $R^{n}$ the function $\phi$ defined by
    $$
    \phi(x)=\frac{1}{2} x^{T} A x-b^{T} x
    $$
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    where the matrix $A$ is symmetric and positive definite. The interest of the question comes from the fact that the minimum of $\phi$ is reached for $x^{*}=A^{-1} b$, and this solution is unique. Indeed, the gradient of $\phi$ in $x=\left(x_{1} \ldots x_{n}\right)^{T} \in R^{n}$ is the vector:
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    $$
    \nabla \phi(x)=\left(\frac{\partial \phi}{\partial x_{1}} \frac{\partial \phi}{\partial x_{2}} \cdots \frac{\partial \phi}{\partial x_{n}}\right)^{T}=\frac{1}{2} A x+\frac{1}{2} A^{T} x-b=A x-b
    $$
</p>

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    which vanishes only for $x^{*}=A^{-1} b$. Note that in this calculation we have used the fact that $A$ is a symmetric matrix. The vector $x^{*}=A^{-1} b$ is therefore the unique critical point of $\phi$ and, since $A$ is positive definite, it is a global minimum.
    <br>
</p>

<p style="font-family: consolas, serif; font-size:14pt; font-style:italic">
    To calculate $x^{(k+1)}$ from $x^{(k)}$, we then minimize the function $\phi$ on the direction vector line $p^{(k)}$ and going through $x^{(k)}$: we therefore choose
    $$
    x^{(k+1)}=x^{(k)}+\alpha_{k} p^{(k)}
    $$
    where $\alpha_{k} \in R$
    <br>
</p>

```python
import numpy as np
```

<p style="font-family: consolas, serif; font-size:12pt; font-style:italic">
    definition of the function $\phi(x)$ and $\nabla \phi(x)$
    <br>
</p>

```python
def phi(A, b, x):
    return 0.5 * np.dot(np.dot(np.transpose(x), A), x) - np.dot(np.transpose(b), x)
```

```python
def Dphi(A, b, x):
    return np.dot(A, x) - b
```

```python
A = np.array([[4, 4, 2, 0], [4, 5, 0, 0], [2, 0, 6, 1], [0, 0, 1, 2]])
B = np.array([1, -1, 0, 1])

n = len(A)
Alpha = 0.01  # step
epsilon = 1e-9  # convergence tolerance
max_iters = 1e6  # maximum iterations

x0 = np.array([1, 0, 0, 1])
```

```python
# Gradient method function
def gradient_methode(A, b, x0):
    # initialization of the starting vector
    next_x = x0
    # Display initial vector
    print("i = 0 ; phi(x)= " + str(phi(A, b, next_x)))
    i = 1
    # convergence flag
    cvg = False
    print("Search start")
    while i <= max_iters:
        curr_x = next_x
        next_x = curr_x - Alpha * Dphi(
            A, b, curr_x
        )  # formula of X(k+1) under the gradient method

        diff = next_x - curr_x
        # convergence test
        if (
            np.linalg.norm(diff, 2) / (np.linalg.norm(next_x, 2) + np.finfo(float).eps)
            <= epsilon
        ):
            cvg = True
            break
        # display the Approximate value of phi(x) during the search for the minimum
        print("i = " + str(i) + " ; phi(x)= " + str(phi(A, b, next_x)))
        i += 1

    # if we have convergence we display the solution as result if not we display a message
    # that the sequence is divergent
    if cvg:
        print("The Min obtained in " + str(i) + " iterations.")
        print("x_sol =", next_x)
    else:
        print("the sequence isn't converge.")

    return next_x
```

<p style="font-family: consolas, serif; font-size:14pt; font-style:italic">
    Execution: <br>** NB: There are +32000 iterations to find the minimum
    <br>
</p>

```python
x_sol = gradient_methode(A, B, x0)
```

```python
print("The solution by the gradient method:")
print("X =  ",x_sol)
```

```python
print("The solution by the classic method:")
x_sol2 = np.linalg.solve(A, B)
print(x_sol2)
```
